```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("paradise.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("image")
plt.axis("off")
plt.show()



```

```python
print(img.shape)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray,cmap="gray")
plt.title("gray scale ")
plt.axis("off")
plt.show()
```

```python
sampled1 = cv2.resize(gray, (200,200))
sampled2 = cv2.resize(gray, (100,100))
sampled3 = cv2.resize(gray, (50,50))


plt.imshow(sampled1, cmap="gray")
plt.title("200x200")
plt.axis("off")
plt.show()

plt.imshow(sampled2, cmap="gray")
plt.title("100x100")
plt.axis("off")
plt.show()

plt.imshow(sampled3, cmap="gray")
plt.title("50x50")
plt.axis("off")

plt.show()
```
