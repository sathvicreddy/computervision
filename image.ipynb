How a Camera Sensor Works

A digital camera sensor contains photodiodes that measure the intensity of light.

Light enters the camera sensor
Incoming photons (light particles) reach the sensor.

Bayer Filter
On top of the sensor there is a Bayer filter which has tiny red, green, and blue color filters.
Each filter allows only one specific color of light to pass through to the photodiode below it.

Photoelectric Effect
When the photons hit the photodiodes, they trigger the photoelectric effect.
This causes the photodiode to release electrons, producing an analog electrical signal proportional to the light intensity.

Analog to Digital Conversion (ADC)
The ADC (Analog-to-Digital Converter) converts this analog signal into digital values.

Image Formation
The digital values represent the intensity of red, green, and blue light, which the processor combines to form the final image.

Quantization in  Image processing

Quantization is the process of converting continuous intensity values into a limited number of discrete levels.

In an 8-bit grayscale image, pixel intensities normally range from:

0 to 255

Example pixel values:
[23, 45, 67, 120, 150, 210]

Quantization Formula
If we quantize using step size = 64, the formula is:
quantized_value = (pixel_value // 64) * 64
This divides the range 0 to 255 into 4 levels.

Example Calculation
Original Pixel	Calculation	Quantized Value
23	(23 // 64) * 64	   =0
45	(45 // 64) * 64	   =0
67	(67 // 64) * 64	   =64
120	(120 // 64) * 64   =64
150	(150 // 64) * 64   =128
210	(210 // 64) * 64   =192

Result:
[0, 0, 64, 64, 128, 192]

Quantization Levels
Because the step size is 64, all values map into 4 possible levels:
0
64
128
192
So many different pixel values get mapped to the same level, which reduces precision but also reduces data size.

Simple Definition

Quantization: Mapping many intensity values into a smaller set of discrete levels.
